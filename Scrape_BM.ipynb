```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://bm.cari.com.my/portal.php"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
#print(soup.text)
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```
